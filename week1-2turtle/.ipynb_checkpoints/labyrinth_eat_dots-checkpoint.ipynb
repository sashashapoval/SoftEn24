```python
from tkinter import*
from random import sample
root = Tk()
# drawing of rectangle
h0 = 20
w0 = 20
side = 20
n_rows = 20 #number of rows
n_clmns = 24 #number of columns

h_rect = side * n_rows
w_rect = side * n_clmns
w = w_rect + 2*w0
h = h_rect + 2*h0
#w = 500
#h = 500

c = Canvas (root, width = w, height = h, background = 'yellow')
c.focus_set()
c.pack()


v = c.create_oval(200, 200, 200, 200, fill = 'yellow')


def ind2center(i, j, side, w0, h0):
    x = w0 + j*side + 0.5*side
    y = h0 + i*side + 0.5*side
    return x, y

def center2ind(x, y, side, w0, h0):
    i = int((y - h0) / side - 0.5)
    j = int((x - w0) / side - 0.5)
    return i, j

for i in range(n_clmns+1): #vertial lines
    c.create_line(w0 + i * side, h0, w0 + i * side, \
                  h0 + h_rect, fill = 'gray')
for i in range(n_rows+1):
    c.create_line(w0, i * side + h0, w_rect + w0, \
                  i * side + h0, fill = 'gray')
   

ball = c.create_oval(w0 + 0.1 * side, h0 + 0.1 * side, w0 +  0.9 * side, \
                     h0 + 0.9 * side, fill = 'red')

#create the list of treasures;
#tr_in_cell[i][j] is index / -1 if the treasury is / is not in the cell:
tr_in_cell = []
for i in range(n_rows):
    tr_in_cell.append([])
    for j in range(n_clmns):
        tr_in_cell[i].append([])
        tr_in_cell[i][j] = -1 # Initially, the treasury is absent
        
#draw treasury
n_treasury = 10 #number of black dots (treasuries)
tmp = sample(range(n_clmns * n_rows), n_treasury) #coordinates (i,j)
# as a single number i * n_rows + j
c4tres = 0.1 #for drawing
tres = [] # list of the black dots
for k in range(len(tmp)):
    j, i = divmod(tmp[k], n_rows)
    x, y = ind2center(i, j, side, w0, h0)
    tres.append(c.create_oval(x - c4tres*side, y - c4tres*side,\
                    x + c4tres*side, y + c4tres*side, fill = 'black'))
    tr_in_cell[i][j] = k


def color_back():
    c.itemconfig(ball, fill = 'red')

def check_tres(coord):
    global n_treasury
    x = (coord[0] + coord[2]) / 2
    y = (coord[1] + coord[3]) / 2
    i, j = center2ind(x, y, side, w0, h0)
    if i >= 0 and i < n_rows and j >= 0 and j < n_clmns:
        if tr_in_cell[i][j] >= 0:
            c.delete(tres[tr_in_cell[i][j]])
            tr_in_cell[i][j] = -1
            n_treasury -= 1
            if n_treasury <= 0:
                c.create_text(w/2, h/2, text = \
                    'Hurrah! You win', font=('Arial', 24, 'bold'))
                root.focus_set()
    else:
        print('Wrong i = ', i, 'j = ', 'x = ', x, 'y = ', y)

def move_up():
    coord = c.coords(ball)
    if coord[1] > h0 + side:
        c.move(ball, 0,  - side)
        coord = c.coords(ball)
        check_tres(coord)
    else: #paint to and backward
        c.itemconfig(ball, fill = 'green')
        c.after(500, color_back)
       

def move_down():
    coord = c.coords(ball)
    if coord[1] > h0 + h_rect - side:
        c.itemconfig(ball, fill = 'green')
        c.after(500, color_back)
    else:
        c.move(ball, 0, side)
        coord = c.coords(ball)
        check_tres(coord)

def move_left():
    coord = c.coords(ball)
    if coord[0] > w0 + side:
        c.move(ball, -side,  0)
        coord = c.coords(ball)
        check_tres(coord)
    else: #paint to and backward
        c.itemconfig(ball, fill = 'green')
        c.after(500, color_back)
       
def move_right():
    coord = c.coords(ball)
    if coord[0] > w0 + w_rect - side:
        c.itemconfig(ball, fill = 'green')
        c.after(500, color_back)        
    else:
        c.move(ball, side, 0)
        coord = c.coords(ball)
        check_tres(coord)
       
c.bind('<Down>', lambda event: move_down())
c.bind('<Up>', lambda event: move_up())
#c.bind('<Up>', lambda event: c.move(ball, 0,  -side))
c.bind('<Right>', lambda event: move_right())  
c.bind('<Left>', lambda event: move_left())


mainloop()
```
